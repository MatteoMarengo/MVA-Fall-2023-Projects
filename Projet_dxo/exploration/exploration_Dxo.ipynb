```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Function to apply selective tone adjustments
def apply_selective_tone(image, highlights, midtones, shadows, blacks):
    # Convert to float for precision in adjustments
    img_float = image.astype(np.float32) / 255.0
    
    # Shadows: Increase the value of the dark pixels
    img_float[img_float < 0.3] *= (1 + shadows)
    
    # Highlights: Decrease the value of the bright pixels
    highlight_mask = img_float > 0.7
    img_float[highlight_mask] = 1 - ((1 - img_float[highlight_mask]) * (1 - highlights))
    
    # Midtones: Adjust mid-range pixels
    midtone_mask = (img_float >= 0.3) & (img_float <= 0.7)
    img_float[midtone_mask] *= midtones
    
    # Blacks: Set the black point
    img_float[img_float < blacks] = 0
    
    # Clip the values to [0, 1] to avoid wrapping around
    img_float = np.clip(img_float, 0, 1)
    
    # Convert back to 8-bit values and return
    img_adjusted = (img_float * 255).astype(np.uint8)
    return img_adjusted

# Load a sample image
image_path = 'D:\OneDrive\Documents\MVA-ENS-2023-2024\S1\IETI\Machine-Intelligence-Images\miniprojectDXO\erasme.jpg'
image = cv2.imread(image_path, cv2.IMREAD_COLOR)

# If the image is in BGR format (as OpenCV loads images), convert it to RGB.
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Apply selective tone adjustments with sample values
# These values are just for demonstration and normally would be set according to user input
highlights = 0.8  # Example value to brighten highlights
midtones = 1.0    # Neutral value for midtones
shadows = 1.2     # Example value to darken shadows
blacks = 0.1      # Example value for black point

# Adjust the image
adjusted_image = apply_selective_tone(image_rgb, highlights, midtones, shadows, blacks)

# Display the original and adjusted images side by side using matplotlib
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(adjusted_image)
plt.title('Adjusted Image')
plt.axis('off')

plt.show()
```

```python
# Let's first define the dodge and burn functions as simple linear transformations.
def dodge(I, strength):
    # This is a simplified version of a dodge operation
    # Typically, the dodge operation lightens the image. The strength parameter controls the intensity.
    # The actual implementation may be different.
    return np.clip(I + strength * I, 0, 255)

def burn(I, strength):
    # This is a simplified version of a burn operation
    # Typically, the burn operation darkens the image. The strength parameter controls the intensity.
    # The actual implementation may be different.
    return np.clip(I - strength * I, 0, 255)

# Assuming I is the input image and we have already normalized it to the range [0, 1]
# For this example, let's create a linear gradient image
I = image1[0, :]/255.0

# Y[p] here is simply the pixel value in a grayscale image, used for masking.
Y = I  # This is the "Highlight Mask"
M_low = 1 - Y  # This is the "Shadow Mask"
M_med = -4 * (Y**2) + 4 * Y  # This is the "Midtone Mask"

# Now apply the mask and operator A (dodge or burn) to the input image
# We'll use a strength factor of 0.5 for demonstration
strength = 0.5
O_dodge = M_med * dodge(I, strength) + (1 - M_med) * I
O_burn = M_med * burn(I, strength) + (1 - M_med) * I

# Plot the results
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))

# Original Image
plt.subplot(2, 3, 1)
plt.plot(image1[0, :]/255.0)
plt.title('Original Image')

# Dodged Image
plt.subplot(2, 3, 2)
plt.plot(O_dodge)
plt.title('Dodged Image')

plt.subplot(2, 3, 3)
plt.plot(image2[0, :, 0]/255.0)
plt.title('DxO Image')

plt.subplot(2, 3, 5)
plt.plot(image3[0, :, 0]/255.0)
plt.title('DxO-1 Image')

# Burned Image
plt.subplot(2, 3, 4)
plt.plot(O_burn)
plt.title('Burned Image')

plt.show()
```
